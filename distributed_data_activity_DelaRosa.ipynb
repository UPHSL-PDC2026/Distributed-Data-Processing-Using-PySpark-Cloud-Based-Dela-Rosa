```python
from pyspark.sql.functions import col

from google.colab import files
files.upload()

df = spark.read.csv("facility.csv", header=True, inferSchema=True)
df.show(5)
```

```python
filtered_df = df.filter(col("Capacity") > 100)
filtered_df.show()
```

```python
grouped_df = df.groupBy("type").count()
grouped_df.show()
```

```python
sorted_df = grouped_df.orderBy("count", ascending=False)
sorted_df.show()
```
